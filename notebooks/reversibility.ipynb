The distance of a Markov chain to beeing reversible can be measured via:
$$\Delta P := \sum_{(i,j)\in\mathcal{A}}\Lambda_s[i,j]X_s[:,j]X_s[:,i]^TD,$$
where $\mathcal{A}=\lbrace (i,j)\in\lbrace 1,\ldots,N\rbrace^2|\ i\neq j,\ \Lambda_s[i,j]\neq 0\rbrace$.
$P-\Delta P$ is reversible and thus, the non-reversibility can be measured by $\Vert\Delta P\Vert$, where $\Vert .\Vert$ is the Frobenius norm for instance.

The following two functions plot the measured non-reversibility depending on the parameters sigma and tau, when using the automated method.

```julia
image = 4
person = 9
mirrored = 0
startPos = 52
tau = 40
sigmin = 2
sigmax = 20

Hokusai.revPerSigma(image, person, tau, mirrored, startPos, sigmin, sigmax);
```

```julia
image = 4
person = 9
mirrored = 0
startPos = 52
sigma = 50
taumin = 2
taumax = 20

Hokusai.revPerTau(image, person, sigma, mirrored, startPos, taumin, taumax);
```

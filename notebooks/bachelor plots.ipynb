```julia
"convert old parameter format to the new one"
function oldtonew!(p)
  p[:, 3:4] = p[:, 4:-1:3]        # adjust for alphabetic order (tau,sigma -> sigma, tau)
  p[:, 3]   = p[:, 3] ./ sqrt(2); # adjust for corrected factor in gaussian computation
  p
end

pba = Any[
# i n t s  
1 5 20 35
1 6 50 60
1 10 60 50

5 6 100 250
5 7 80 120
5 9 80 120
    
7 6 80 80 
7 8 80 80
7 10 80 100] |> oldtonew!

for i = 1:size(pba, 1)
  srand(1)
  Hokusai.run(pba[i,:]..., precluster=pcl)
end
```
